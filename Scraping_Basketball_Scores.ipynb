```python
years = list(range(1991, 2025))
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
# Télécharger et sauvegarder les pages HTML
for year in years:
    url = url_start.format(year)
    data = requests.get(url)
    # Utilisation de l'encodage 'utf-8' pour écrire les fichiers
    with open(f"mvp/{year}.html", "w+", encoding="utf-8") as f:
        f.write(data.text)

print("Téléchargement terminé, fichiers sauvegardés dans le dossier 'mvp'.")
```

# NBA Player Stats: Per Game 

```python
import os
import requests

# Créer le dossier 'player' s'il n'existe pas
if not os.path.exists("player"):
    os.makedirs("player")

player_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"

for year in years:
    url = player_stats_url.format(year)
    
    data = requests.get(url)
    
    # Enregistrer la page HTML dans le dossier 'player'
    with open(f"player/{year}.html", "w+", encoding="utf-8") as f:
        f.write(data.text)

print("Téléchargement terminé, fichiers sauvegardés dans le dossier 'player'.")
```

```python
dfs = []
for year in years:
    with open(f"player/{year}.html", encoding="utf-8") as f:  # Spécifiez l'encodage
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    
    # Supprimer l'en-tête redondant si présent
    if soup.find('tr', class_="thead"):
        soup.find('tr', class_="thead").decompose()
    
    # Extraire le tableau des statistiques
    player_table = soup.find(id="per_game_stats")
    if player_table:
        player_df = pd.read_html(str(player_table))[0]
        player_df["Year"] = year
        dfs.append(player_df)
    else:
        print(f"Aucun tableau trouvé pour l'année {year}.")
```

```python
players = pd.concat(dfs)
```

# Teams_Stats

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
# Créer le répertoire "team" s'il n'existe pas
os.makedirs("team", exist_ok=True)

for year in years:
    url = team_stats_url.format(year)
    
    data = requests.get(url)
    
    # Sauvegarder le contenu dans un fichier HTML
    with open(f"team/{year}.html", "w+", encoding="utf-8") as f:
        f.write(data.text)
    print(f"Page pour l'année {year} sauvegardée avec succès.")
```

```python
dfs = []
for year in years:
    with open("team/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    
    # Vérification si l'élément avec la classe "thead" existe
    thead = soup.find('tr', class_="thead")
    if thead:
        thead.decompose()  # Supprimer l'élément si trouvé
    
    # Extraire les tableaux des conférences Est et Ouest avec vérification
    e_table = soup.find_all(id="divs_standings_E")
    if e_table:
        e_df = pd.read_html(str(e_table[0]))[0]
        e_df["Year"] = year
        e_df["Team"] = e_df["Eastern Conference"]
        del e_df["Eastern Conference"]
        dfs.append(e_df)
    
    w_table = soup.find_all(id="divs_standings_W")
    if w_table:
        w_df = pd.read_html(str(w_table[0]))[0]
        w_df["Year"] = year
        w_df["Team"] = w_df["Western Conference"]
        del w_df["Western Conference"]
        dfs.append(w_df)

```

```python
teams = pd.concat(dfs)
```
