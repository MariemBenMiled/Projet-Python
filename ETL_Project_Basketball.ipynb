```python

def update_team_column(dataframe):
    """
    Met à jour la colonne 'Team' du DataFrame pour remplacer les valeurs '2TM', '3TM', '4TM', '5TM' par 'TOT'.
    
    Args:
        dataframe (pd.DataFrame): Le DataFrame contenant la colonne 'Team'.
    
    Returns:
        pd.DataFrame: Le DataFrame mis à jour.
    """
    try:
        # Vérifier si la colonne 'Team' existe dans le DataFrame
        if 'Team' in dataframe.columns:
            # Remplacer les valeurs spécifiques par 'TOT'
            dataframe['Team'] = dataframe['Team'].replace(['2TM', '3TM', '4TM', '5TM'], 'TOT')
            print("Les valeurs de la colonne 'Team' ont été mises à jour avec succès.")
        else:
            print("Erreur : La colonne 'Team' n'existe pas dans le DataFrame.")
    except Exception as e:
        print(f"Erreur lors de la mise à jour de la colonne 'Team' : {e}")
    
    return dataframe

# Exemple d'utilisation avec votre DataFrame players_stats_df
players_stats_df = update_team_column(players_stats_df)

# Afficher les premières lignes pour vérifier le résultat
print(players_stats_df.head())
```

```python
def remove_columns(dataframe, columns_to_remove):
    """
    Supprime les colonnes spécifiées d'un DataFrame.

    Args:
        dataframe (pd.DataFrame): Le DataFrame à modifier.
        columns_to_remove (list): Liste des colonnes à supprimer.

    Returns:
        pd.DataFrame: Le DataFrame sans les colonnes spécifiées.
    """
    try:
        # Vérifier si les colonnes à supprimer existent dans le DataFrame
        existing_columns = [col for col in columns_to_remove if col in dataframe.columns]
        
        if existing_columns:
            # Supprimer les colonnes spécifiées
            dataframe = dataframe.drop(columns=existing_columns, axis=1)
            print(f"Les colonnes suivantes ont été supprimées avec succès : {existing_columns}")
        else:
            print(f"Aucune des colonnes spécifiées n'existe dans le DataFrame : {columns_to_remove}")
    except Exception as e:
        print(f"Erreur lors de la suppression des colonnes : {e}")
    
    return dataframe

# Liste des colonnes à supprimer
columns_to_remove = ['Unnamed: 0', 'Awards']

# Appliquer la fonction à votre DataFrame
players_stats_df = remove_columns(players_stats_df, columns_to_remove)

# Afficher les premières lignes pour vérifier le résultat
print(players_stats_df.head())
```

```python
# 1. Insérer les données dans `dim_player`
def insert_into_dim_player(df, engine):
    # Sélectionner les colonnes nécessaires
    player_data = df[['player_name', 'player_age', 'position']].drop_duplicates()
    
    # Insérer dans SQL Server
    try:
        player_data.to_sql('dim_player', con=engine, if_exists='append', index=False)
        print("Données insérées dans dim_player.")
    except Exception as e:
        print(f"Erreur lors de l'insertion dans dim_player : {e}")

# 2. Insérer les données dans `dim_team`
def insert_into_dim_team(df, engine):
    # Sélectionner les colonnes nécessaires
    team_data = df[['team_name', 'full_team_name']].drop_duplicates()
    
    # Insérer dans SQL Server
    try:
        team_data.to_sql('dim_team', con=engine, if_exists='append', index=False)
        print("Données insérées dans dim_team.")
    except Exception as e:
        print(f"Erreur lors de l'insertion dans dim_team : {e}")

# 3. Insérer les données dans `dim_season`
def insert_into_dim_season(df, engine):
    # Sélectionner les colonnes nécessaires
    season_data = df[['season_year']].drop_duplicates()
    
    # Insérer dans SQL Server
    try:
        season_data.to_sql('dim_season', con=engine, if_exists='append', index=False)
        print("Données insérées dans dim_season.")
    except Exception as e:
        print(f"Erreur lors de l'insertion dans dim_season : {e}")

# 4. Appeler les fonctions d'insertion
insert_into_dim_player(stats, engine)
insert_into_dim_team(stats, engine)
insert_into_dim_season(stats, engine)
```

```python
# 1. Insérer les données dans `fact_stats`
def insert_into_fact_stats(df, engine):
    """
    Insère les données dans la table fact_stats en se basant sur les clés étrangères des tables de dimension.
    """
    try:
        # Mapper les IDs des dimensions pour chaque ligne
        query_player = "SELECT player_id, player_name FROM dim_player;"
        query_team = "SELECT team_id, team_name FROM dim_team;"
        query_season = "SELECT season_id, season_year FROM dim_season;"
        
        # Charger les données des dimensions
        with engine.connect() as conn:
            dim_player = pd.read_sql(query_player, conn)
            dim_team = pd.read_sql(query_team, conn)
            dim_season = pd.read_sql(query_season, conn)
        
        # Mapper player_id
        df = df.merge(dim_player, left_on='player_name', right_on='player_name', how='inner')
        
        # Mapper team_id
        df = df.merge(dim_team, left_on='team_name', right_on='team_name', how='inner')
        
        # Mapper season_id
        df = df.merge(dim_season, left_on='season_year', right_on='season_year', how='inner')
        
        # Sélectionner les colonnes nécessaires pour la table des faits
        fact_data = df[[
            'player_id', 'team_id', 'season_id', 
            'games_played', 'minutes_played', 'field_goals_made',
            'field_goals_attempted', 'field_goal_percentage', 'threepoint_made',
            'points_scored', 'rebounds', 'assists', 'steals', 'blocks', 
            'wins', 'losses', 'win_loss_percentage', 
            'points_scored_per_game', 'points_allowed_per_game', 
            'simple_rating_system', 'games_started'
        ]]
        
        # Insérer dans SQL Server
        fact_data.to_sql('fact_stats', con=engine, if_exists='append', index=False)
        print("Données insérées dans fact_stats.")
    
    except Exception as e:
        print(f"Erreur lors de l'insertion dans fact_stats : {e}")

# 2. Appeler la fonction pour insérer les données dans la table des faits
insert_into_fact_stats(stats, engine)
```
